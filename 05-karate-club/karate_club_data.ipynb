```python
import networkx as nx
```

```python
data = nx.karate_club_graph()
```

```python
nx.draw_networkx(data)
```

```python
data.number_of_nodes()
```

```python
data.number_of_edges()
```

```python
club_list = []
for i in range(data.number_of_nodes()):
    club_names = data.nodes[i]['club']
    print(club_names)
    # club_list.append([0 if club_names == "Mr. Hi" else 1])
    if club_names == "Mr. Hi" :
        club_list.append(0)
    elif club_names == "Officer":
        club_list.append(1)

club_list
```

```python
nx.draw_networkx(data, node_color = club_list, cmap='coolwarm')
```

```python
nx.draw_circular(data, node_color = club_list, cmap='coolwarm')
```
