```python
import dgl
import networkx as nx
import numpy as np
import torch
```

```python
dataset = dgl.data.CiteseerGraphDataset()
dataset
```

```python
G = dataset[0]
G
```

```python
G_nx = dgl.to_networkx(G, node_attrs=['label', 'feat'], edge_attrs=None)
```

```python
G_nx.nodes(data=True)
```

```python
import matplotlib.pyplot as plt
```

```python
options = {
    'node_color':'red',
    'node_size':1,
    'width':0.02,
    'edge_colors':'gray'
}
nx.draw_circular(G_nx, **options, with_labels=False)
# nx.draw_spring()
# nx.draw_networkx_nodes()
# nx.draw_networkx_edges()
plt.show()
```
