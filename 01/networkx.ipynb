# Introduction To NetworkX

```python
import networkx as nx
```

```python
### define Graph G

G = nx.Graph()
G.add_nodes_from([1, 2, 3, 4])
G.add_edges_from([(1,2), (1, 3), (2,3), (2, 4), (2, 5)])
nx.draw(G)
```

```python
# Nodes - vertices
G.nodes
```

```python
G.number_of_nodes()
```

```python
# edges
G.edges
```

```python
G.number_of_edges()
```

```python
G.graph
```

```python
# export graph :

from networkx.readwrite import json_graph
```

```python
# export data(in json)
json_data = json_graph.node_link_data(G)
json_data
```

```python
# export graph
data_to_graph = json_graph.node_link_graph(json_data)
data_to_graph
```

```python
nx.draw(data_to_graph)
```

```python
G.is_directed()
```

```python
G.is_multigraph()
```

```python
# number og nodes :
G.order()
```

```python
# Density of Graph (?)
from networkx.classes.function import density
density(G)
```

```python
# Degree of Graph
from networkx.classes.function import degree
degree(G)
```

```python
from networkx.classes.function import degree_histogram
degree_histogram(G)

# frequency of degree
```

```python
# Neighbors :
from networkx.classes.function import neighbors

for node in G.nodes():
    print(f"Node : {node} is Neighbor with : {list(neighbors(G, node))}")
```

### Directional Graph

`nx.DiGraph()`

```python
di_graph = nx.DiGraph()

di_graph.add_edges_from([(1, 2), (2, 3)])

nx.draw(di_graph, with_labels=True)
```

```python
nx.is_directed(di_graph)
```

```python
nx.is_weighted(di_graph)
```

```python
di_graph.nodes
```

## Neighbor & Adjacency

```python
G1 = nx.DiGraph()

G1.add_edge('a', 'b', weight = 5)
G1.add_edge('a', 'c', weight = 2)
G1.add_edge('d', 'c', weight = 3)
G1.add_edge('f', 'c', weight = 3)
G1.add_edge('f', 'a', weight = 3)

nx.draw(G1, with_labels=True)
```
